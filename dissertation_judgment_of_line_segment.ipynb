```python
import numpy as np
```

```python
R = [[(0, 0), (1, 1)], [(1.9,2.8), (5.6,6.6)], [(2.8,2.8), (4.8, 5)], [(0.75,0.75), (0.5, 0.5)], [(0, 0), (0.24, 0.24)], [(0, 0), (0.9, 0.9)]]
R
```

```python
R = np.array(R)
R
```

```python
def judgement(R, A, Tot):
    for i in range(len(R)):
        Tot = Tot + 1
        if R[i][0][0] < R[i][1][0]:
            start = R[i][0]
            end = R[i][1]
        elif R[i][0][0] > R[i][1][0]:
            end = R[i][0]
            start = R[i][1]
        else:
            if R[i][0][1] < R[i][1][1]:
                start = R[i][0]
                end = R[i][1]
            else:
                end = R[i][0]
                start = R[i][1]

        A[0] += (start - A[0]) / Tot
        A[1] += (end - A[1]) / Tot
        A = np.array([A[0], A[1]])
    return A, Tot
```

```python
Tot = 0
A = [(0, 0), (0, 0)]
i, p = 0, 1

while p <= len(R):
    Res = R[i:p]
    A, Tot = judgement(Res, A, Tot)
    print(f"After {p}th response : start point {A[0]} and end point {A[1]}")
    i = p
    p += 1
```
