```python
def swap(p, q):
    return q.copy(), p.copy()
```

```python
def valid_length(R, max_len):
    distances = np.linalg.norm(R[:, 0] - R[:, 1], axis=1)
    mask = distances <= max_len

    List = R[mask]
    IndexList = np.where(mask)[0]

    return List, IndexList
```

```python
def valid_distance(List, IndexList, i, j, min_dis):
    d1 = math.dist(List[i][0], List[j][0])
    if d1 < min_dis:
        List = np.delete(List, j, axis = 0)
        IndexList = np.delete(IndexList, j)
    else:
        d2 = math.dist(List[i][0], List[j][1])
        if d2 < min_dis:
            List = np.delete(List, j, axis = 0)
            IndexList = np.delete(IndexList, j)
        else:
            d3 = math.dist(List[i][1], List[j][0])
            if d3 < min_dis:
                List = np.delete(List, j, axis = 0)
                IndexList = np.delete(IndexList, j)
            else:
                d4 = math.dist(List[i][1], List[j][1])
                if d4 < min_dis:
                    List = np.delete(List, j, axis = 0)
                    IndexList = np.delete(IndexList, j)
                else:
                    j = j + 1
    return List, IndexList, j
```

```python
def find_mean(R, A, Tot, freq):
    A[0] += freq * (R[0] - A[0]) / Tot
    A[1] += freq * (R[1] - A[1]) / Tot
    A = np.array([A[0], A[1]])
    return A
```

```python
#R is the response of one annotator, A is the array of previously stored points
def find_multiple_lines(R, A, Tot, min_dis, max_len): 
    #Sorting line segment
    for i in range(len(R)):
        if R[i][0][0] > R[i][1][0]:
            R[i][0], R[i][1] = swap(R[i][0], R[i][1])
        elif R[i][0][0] == R[i][1][0]:
            if R[i][0][1] > R[i][1][1]:
                R[i][0], R[i][1] = swap(R[i][0], R[i][1])
            
    for i in range(len(R)):
        for j in range(i + 1, len(R)):
            if R[i][0][0] > R[j][0][0]:
                R[i], R[j] = swap(R[i], R[j])
            elif R[i][1][0] == R[i][1][0]:
                if R[i][0][1] > R[j][0][1]:
                    R[i], R[j] = swap(R[i], R[j])


    #Length check for valid entry
    List, IndexList = valid_length(R, max_len)
    # R = None

    #Distance check for valid entry
    i = 0
    while i < (len(List)):
        j = i + 1
        while j < len(List):
            List, IndexList, j = valid_distance(List, IndexList, i, j, min_dis)
        i = i + 1

    #Find mean
    freq = len(List)
    for i, index in enumerate(IndexList):
        Tot[index] += freq    #Put 'freq' times importance of the response

    # List = freq * List    #Put 'freq' times importance of the response
    for i, index in enumerate(IndexList):
        A[index] = find_mean(List[i], A[index], Tot[index], freq)

    return A, Tot
```

```python
def make_response(s):
    Row_list = []
    for i in range(3):
        a = list(map(float, s[2*i+1].split(',')))
        b = list(map(float, s[2*i+2].split(',')))
        Row_list.append([a, b])
    return Row_list    
```
